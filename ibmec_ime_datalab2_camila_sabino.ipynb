```python
# Definindo vetor para armazenar uma série histórica
count = 25
#print("Minha variável: ", count)

anos_observados = [0] * count
ano_inicial = 2000

# Preenche o vetor que armazena a série histórica
for x in range(count):
  anos_observados[x] = ano_inicial + x

# print(anos_observados)

# Seleciona no vetor os anos desejados
anos_selecionados = [0] * 15


# Preenche o vetor dos anos selecionados
i = 0
for x in anos_observados:
    if x >= 2010:
        anos_selecionados[i] = x
        i += 1


# Preenche o vetor de valores que eu desejo observar na série histórica
valores_selecionados = [7.5, 4.0, 1.9, 3.0, 0.5, -3.5, -3.3, 1.3, 1.8, 1.1, -4.1, 4.6, 3.1, 3.2, 3.5]

# Verifica os valores criados para a série histórica
print(anos_selecionados)
print(valores_selecionados)
```
